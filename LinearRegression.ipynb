```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('salary_linearreg.csv')
```

```python
df
```

```python
df.describe()
```

```python
X = df[['YearsExperience']]  # X should be a 2D array
Y = df['Salary']    # Y is usually a 1D array
```

```python
plt.scatter(X,Y)
plt.xlabel('YearsExperience', fontsize = 20)
plt.ylabel('Salary', fontsize = 20)
plt.title('Scatter plot of Experience Years vs Salary')
plt.show()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(X, Y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.show()
```

```python
print(f'Coefficient: {model.coef_[0]}')
print(f'Intercept: {model.intercept_}')
```
