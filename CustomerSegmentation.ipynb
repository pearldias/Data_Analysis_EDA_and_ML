```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("Mall_Customers.csv")
print(df.head())
```

```python
X = df[["Annual Income (k$)", "Spending Score (1-100)"]]
```

```python
# Apply K-Means with 3 clusters (you can adjust this)
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x=df["Annual Income (k$)"], y=df["Spending Score (1-100)"], hue=df["Cluster"], palette="viridis", s=100)
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation (K-Means)")
plt.legend(title="Cluster")
plt.show()
```

```python
#Interpretation: 
#Cluster 0: High Income, High Spending (Luxury Shoppers)
#Cluster 1: Low Income, Low Spending (Budget Customers)
#Cluster 2: Average Income, Moderate Spending (Casual Shoppers))
```
